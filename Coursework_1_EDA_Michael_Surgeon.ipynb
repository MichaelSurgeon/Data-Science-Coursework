## Common issues within data

Data in the real world can have a host of issues that can cause problems during analysis. We have to think about how to best handle these issues to provide non bias and accurate analysis. These issues include but are not limited to, duplicates, missing data, outliers and human error.

### Duplicates
Duplicates, are rows that are repeated throughout data. It's important to identify duplicate records, as duplicate entries can lead to inaccurate results, by inflating certain features particularly frequency columns potentially causing bias i.e.a duplicated country could bias analysis towards it. We must identify these duplicates and remove them from the dataset or address them appropriately if deemed useful.


### Missing data
Missing data refers to missing values within a given record i.e. take the Social Support feature, this might have NA (Not available), NaN (Not a number) or just blank within a given entry.

Missing data comes in different forms, each comes with their own risk of bias. Meaning if the data is missing there could be potentialy a bias in the data i.e. people who dont report high salaries in a dataset means the dataset is now biased towards low earners. Types of missing data are categorised into:

- MCAR (Missing completely at random)
  - There is no pattern within the data (on any feature) as to why the data is missing.
  - Risk Bias - Low

- MAR (Missing at random)
  - Missing data is related to observed data i.e. when one column is missing another column is.
  - Risk Bias: Moderate

- MNAR (Missing not at random)
  - Missing data, but related to the feature its missing from for example  people might not want to report their age or income.
  - Risk Bias: High (Can greatly affect overall data accuracy)

When we have missing data, we have to figure out how best to handle it. Typically we handle missing data using a technique called imputation, which is a technique used to fill in missing data. There are a few types of imputation we can perform:

- Mean Imputaion - Fill with the average value accross a certain feature
- Median Imputation - Fill the missing value with the exact middle between the data point.
- Mode imputation - Use the most frequent value from the feature.
- Forward fill - Use the last value
- Predictive - Use a model to predict the value.

### Inconsistent Data
Inconsistent data is a typical issue we seen within real world data. Inconsistent formatting i.e. dates, number respresenation etc can cause various problems during later analysis such as parsing issues, un-reliable data & data types.

For example:

- Dates
  - mm/dd/yyyy vs dd-mm-yyyy (can cause parsing issues)
- Number format
  - 10K vs 10,000 vs 10_000 vs 10000
- Data Type
  - "10" vs 10

we should, through data cleansing, standardised data as it equates to a better experince when analysing the data. Along with providing stronger insights.

### Outliers
Outliers are described as values that deivate drastically from the "normal" i.e. average salary of 50k and outlier of 500k.

Outliers can potentially skew data deviating from a "normal" distribution, or cause bias. However, they can be useful if used correctly & can provide very valuable insights as to why a value is particularly high.

There are a few methods we can use for outlier detection these include:

- Visual inspection - we can use things like boxplots or histograms to show outliers.
- Z-Score - measure how many standard devisations a particular dataset is from its mean. We can then set a threshold for outliers.
- InterQuartile range (IQR) - identifies the spread of data within 50% of the data, anything outwith is an outlier. IQR is the upper quartile limit Q3 minus the lower limit Q1.
- Domain knowledge - If you know a particular domain i.e. how much what is a very high level of corruption, you can set a custom threshold and flag any outliers above that threshold.

### Outliers caused by human error
Human error is something we must be cautious of within real world data. For example, a typo of a ladder score value can cause outliers, or typos effecting mode imputation, this can lead to bias. We have to be aware of the data we are working within and address issues where applicable.

### Outliers

Outlier detection was done via z-score detection with a standard deviation threshold of 3.

It showed some intresting outliers, noticeably Venezuela which had a 0 Log GDP per captia which is not plausible.

Through further analysis another intresting set of outliers were the freedom & social support features of Afghanistan. Unlike Venezuela, these could be valid values, it is highly plasubile that there is a country without social support or freedom, thus they were kept in the dataset.

```python

df_clean_outliers = df_clean_features.copy()

print("Outliers in data per feature")
print("=============")
from scipy.stats import zscore
import matplotlib.pyplot as plt
import seaborn as sns

# plot outliers on numeric cols for visual rep
numeric_cols = ['Ladder score', 'Log GDP per capita', 'Social support', 'Freedom to make life choices', 'Perceptions of corruption']

plt.figure(figsize=(14,5))
sns.boxplot(data=df_clean_outliers[numeric_cols])
plt.title('Boxplot per feature to identify outliers.')
plt.show()

cols_to_check = df_clean_outliers.columns.tolist()
cols_to_check.remove("Country name")

print("\nOutliers by > 3 threshold")
for col in cols_to_check:
    df_clean_outliers["z_score"] = zscore(df_clean_outliers[col], nan_policy="omit")
    outliers = df_clean_outliers[df_clean_outliers["z_score"].abs() > 3]

    if not outliers.empty:
        print(f"\nOutliers in column: {col}")
        for idx, row in outliers.iterrows():
            print(f"  Row {idx} - Country: {row['Country name']} - Value: {row[col]}")
    else:
        print(f"\nNo outliers in column: {col}")

# remove to not affect further analysis
df_clean_outliers = df_clean_outliers.drop(columns=['z_score'])
print("\n")
```

## Methods Used

In order to better understand a features distrubtion, central tendancy & spread along with gaining key insights. Within this section of analysis I used the following methods:

- Summary Statistics
  - **What it is**: Numeric statistics of a given feature that shows mean, min, max, percentiles & the average standard deviation between the numerical values.
  - **Application**: Used to better understand a features central tendency & spread.

- Value Counts
  - **What it is**: List of unique values in a particular feature i.e.
  - **Application**: Shows spread of data based on number of unique values

- Histogram
  - **What it is**: A plot that groups data into a number of bins (vertical bars).
  - **Application**: Shows data trends/patterns, if data is right or left skewed & overall distribution.

- Box plot
  - **What it is**: - A plot diagram that includes a box, upper and lower limits. Anything outwith these limits are considered outliers.
  - **Box**: represents the interquartile range, the middle 50% of the data. This is calculated by (Q3 - Q1) within the box the line is the median and either side represents 25% of the data.
  - **Upper Whisker**: the line on the top of the box that represents the upper limit (last upper data point), calculated by (Q3 + 1.5 * IQR)
  - **Lower Whisker**: the line on the bottom of the box that represents the lower limit (last lower data point), calculated by (Q1 - 1.5 * IQR)
  - **Application**: Shows a features distribution, spread & central tendency along with key visual outliers, easier to identify than plain numbers.

# Bivariate Analysis

This section will analyse the relationships related to my hypotheses to better understand if they stand true or not, as a reminder here are the hypotheses:

## Hypothesis 1: Freedom & Corruption

In this section I will be testing my hypothesis `Greater freedom and lower corruption results positively impact a countries happiness score` through cross tabulation & pivot table analysis to see if countries with lower corruption and higher freedom are typically happier.

### Cross Tabulation Analysis

Cross tabulation tells you the frequency distribution accross n amount of categorical variables, this can be represented as plain counts or percentage values (distribution of different categories per row combination i.e. low freedom & high corruption = 16.67% of countrys).  

In order to perform cross tabulation analysis I created two new features:

- **Freedom_cat**
  - Splits the freedom scores into low, medium and high based on 3 percentiles.

- **Corruption_cat**
  - Splits the corruption scores into low, medium and high based on 3 percentiles.

I then created a cross tabulation table, in which the rows were both categories and the columns were the ladder score category (low, medium, high), this allowed me to measure what percentage of countries overall had the highest happiness with differing levels of corruption and freedom.

If we look at the following:

- **High Freedom / Low Corruption**
  - 50% of countries are within the high category.
- **High Freedom / Medium Corruption**
  - 66.7% of countries are within the high category.
- **High Freedom / High Corruption**
  - 88% of countries are within the high category.

This analysis was interesting as it provides some support to the first part of my hypothesis, we do typically see higher levels of happiness when there is higher freedom levels. However corruption has a negligible effect on happiness. Shown by the High/High grouping, in this combination the happiness was actually superior than the High/Medium or low grouping.

This established a strong trend in which freedom does have a direct impact on happiness within a country but corruption does not.

```python
df_freedom_corruption = df_ladder_score.copy()

# create new freedom_cat feature
df_freedom_corruption['Freedom_cat'] = pd.qcut(
    df_freedom_corruption['Freedom to make life choices'],
    q=3, # split into 3 quartile bins (Low / Medium / High)
    labels=['Low', 'Medium', 'High'],
)

# create new corruption_cat feature
df_freedom_corruption['Corruption_cat'] = pd.qcut(
    df_freedom_corruption['Perceptions of corruption'],
    q=3, # split into 3 quartile bins (Low / Medium / High)
    labels=['Low', 'Medium', 'High']
)

# create cross tabulation with percentages
cross_tab = pd.crosstab(
    [df_freedom_corruption['Freedom_cat'], df_freedom_corruption['Corruption_cat']],
    df_freedom_corruption['Ladder_score_cat'],
    normalize='index'
) * 100

# remove header name (ladder score)
cross_tab.columns.name = None

# Flatten  multi-level index to turn row labels into columns for better viewing
cross_tab_flat = cross_tab.reset_index().round(2)

print("Proportion of Happiness Levels by Freedom and Corruption")
print("========================\n")
print(cross_tab_flat)
```

### Pivot Table Analysis

I then used a pivot table to futher analyse the relationship between corruption and freedom on happiness.

A pivot table allows us to group data into categories and perfom some sort of aggregation on it like mean, average min max.

Within my pivot table we group the freedom category into the rows, then the corruption category into the coluns so we can see combinations like (low freedom, low corruption) (high freedom, low corruption) etc. The values in the middle are aggregated on the mean of the ladder score feature for a given category.

This analysis showed that countries with high freedom typically have a higher happiness, however corruption does not have much influence on a countries happiness.


```python
pivot = df_freedom_corruption.pivot_table(
    index='Freedom_cat',
    columns='Corruption_cat',
    values='Ladder score',
    aggfunc='mean',
    observed=False
)

pivot = pivot.round(2)

pivot.index.name = 'Freedom Category'
pivot.columns.name = 'Corruption Category'

print("Average Happiness Scores By Freedom and Corruption Levels")
print("========================\n")
print(pivot)
```

### Happiness Levels by Freedom and Corruption (Bar Chart)

To better visualise the findings from the cross tabulation table, I have created a bar chart this shows the correlation for each combination of freedom and corruption.

This further strengthens my stance from the cross tabulation & pivot table analysis.

```python

labels = "F-" + cross_tab_flat['Freedom_cat'].astype(str) + ' / ' + "C-" + cross_tab_flat['Corruption_cat'].astype(str)

fig, ax = plt.subplots(figsize=(8,6))

ax.bar(labels, cross_tab_flat['Low'], label='Low Happiness')
ax.bar(labels, cross_tab_flat['Medium'], bottom=cross_tab_flat['Low'], label='Medium Happiness')
ax.bar(labels, cross_tab_flat['High'], bottom=cross_tab_flat['Low']+cross_tab_flat['Medium'], label='High Happiness')

ax.set_ylabel('Percentage of Countries')
ax.set_title('Happiness Levels by Freedom and Corruption')
ax.legend()

plt.xticks(rotation=45)
plt.tight_layout()

print("Happiness Levels by Freedom and Corruption")
print("========================\n")
plt.show()
```

### Summary

I stated that `Greater freedom and lower corruption results positively impact a countries happiness score`, it's true that greater freedom links directly to happier countries. This was evident in the cross tabulation, where 88% of countries that both reported high freedom and corruption were happier overall within the pivot table, where the average happiness for the same combination was 6.65. The data visualisation within the bar chart showed this well.

I could not state the same about corruption it would appear that no matter the level of corruption, if there is a perception of more freedom a country is generally happier.

This suggests that even reducing corruption does not inherently impact happiness, people are more focused on having a strong sense of freeedom more so than what level of corruption a goverment or organisation has. Based on this analysis, freedom is one of the key factors that effect happiness within a country.

Through further analysis it would be intresting to see if peoples perception on corruption directly correlates to freedom i.e. if people think their goverment introduces unfair tax rises or laws for example they might feel less free, this could be done using a prediction model to verify if higher freedoms predict higherh happiness over time.

## Hypothesis 2: Economic Stance

In this section I look into my second hypothesis `Lower economic states result in a country having a lower happiness score`.

In this section I will use a correlation matrix, and scatter plot to see if there is a direct correlation between GDP per person to a countrys happiness.

### Correlation Matrix

A correlation matrix works by using the pearson correlation formula it shows us the correlation coefficents between a set of variables. It allows us to better understand if relationships are positive (linear) or negative. This is measured between:

- 1 : Linearly Positive
- -1 : Linearly Negative

A positive relationship suggests two or more variables are strongly linked whereas negative relationships suggest the opposite.

The correlation matrix in the case of GDP and happiness indicates a relatively strong association where they have a linear relationship of 0.798 between log GDP and ladder score (happiness), this is a strong indication that my hypothesis is correct.

```python
df_gdp_score = df_ladder_score.copy()

economic_state_corr_data = {
    'Log GDP': df_gdp_score["Log GDP per capita"],
    'Ladder score': df_gdp_score["Ladder score"]
}

df = pd.DataFrame(economic_state_corr_data)

correlation_matrix = df.corr()

print("Correlation Matrix - Log GDP / Ladder Score")
print("========================\n")
print(correlation_matrix)
```

### Correlation Matrix Heatmap

To represent the correlation matrix visually, I have used a correlation matrix heatmap where blue is a more positive assocation and red is negative. Here we can clearly the see the association between the two variables is strong.


```python

print("Correlation Matrix Heatmap")
print("========================")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

### Scatter Plot

I used a scatter plot as another method of visualisation to show the relationship between log GDP and ladder score. A scatter plot, places dots within a graph, each position represents its position on an x and y axis where x and y are data points.

Within this scatter plot its clear to see that as the ladder score increases so does the log gdp. This indicates a positive relationship between the two variables and is a strong indication that higher log GDP does indeed correlate directly to higher overall happiness.

```python
plt.scatter(economic_state_corr_data['Ladder score'], economic_state_corr_data['Log GDP'])

plt.xlabel('Ladder Score')
plt.ylabel('Log GDP')
plt.title('Correlation between Ladder Score & Log GDP')

print("Scater plot - Log GDP / Ladder Score")
print("========================\n")
plt.show()
```

### Summary

The analysis above was related to my hypothesis `Lower economic states result in a country having a lower happiness score`.

We could see from the correlation matrix that there was a strong association between log gdp and the ladder score (happiness) it measured out at 0.80 which is close to being a fully positive relationship.

Additonally, through the scatter plot we seen that as the ladder score grows so does the log GDP. There was a strong association between the two.

This makes sense, if people make more money they can ultimately do what they want, have more financial security which is a driver in a persons life satisfaction. It's no surprise that money is a driver that impacts happiness per person having a significant effect on a country's overall rating.
